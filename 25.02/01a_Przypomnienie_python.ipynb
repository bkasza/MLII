# Pętle explicite vs implicite

**Proszę:**

* stworzyć tablicę `x` o kształcie (n,n) wypełnioną zerami dla `n=100`
* wypełnić w pętli tablicę liczbami losowymi z rozkładu płaskiego w zakresie `[0,1]`
* w kolejnej pętli obliczyć (bez korzystania z funkcji pakietu numpy) sumę elementów: `s` oraz sumę kwadratów elementów: `s2`
* korzystając z `s` oraz `s2` obliczyć wariancję `v` elementów tablicy `x`:
\begin{equation}
V = E_{x\sim p(x)}(x^{2}) - E_{x \sim p(x)}(x)^{2}
\end{equation}
* wypisać na ekran `v*12` (jakiej wartości się spodziewamy?)
* zmierzyć czas wykonywania kodu komórki używając polecenia `time` Jupyter magic na początku komórki:
  
  ```Python
  %%time
  ```

````python
%%time
n = 100
x = np.zeros((n,n)) # ew np.full((n,n), 0.0)
#algorytm kongruentny - najprostszy do generowania liczb pseudolosowych
for i in range(n):
  for j in range(n):
    x[i, j] = np.random.default_rng().uniform(0.0, 1.0) #ew random.default_rng().random()
s, s2 = 0, 0
for i in range(n):
  for j in range(n):
    s += x[i, j]
    s2 += x[i, j]**2
v = s2/x.size - (s/x.size)**2
print(colored("Variance x 12:","blue"),v*12)
````

**Proszę:**

Używając jedynie funkcji biblioteki numpy **bez pętli, z jedną linią, na polecenie**:

* stworzyć tablicę `x` o kształcie (n,n) wypełnioną liczbami losowymi z rozkładu płaskiego w zakresie `[0,1]`
  
* obliczyć sumę elementów: `s` oraz sumę kwadratów elementów: `s2`
  
* korzystając z `s` oraz `s2` obliczyć wariancję `v` elementów tablicy `x`
  
* wypisać na ekran `v*12` (jakiej wartości się spodziewamy?)
  
* zmierzyć czas wykonywania kodu komórki
  

**Proszę:** zwrócić uwagę na czas wykonania obu komórek  

````python
%%time

x = np.random.default_rng().uniform(0.0, 1.0, (n,n))
s = np.sum(x)
s2 = np.sum(x*x)
v = s2/x.size - (s/x.size)**2
print(colored("Variance x 12:","blue"),v*12)
````

# Generacja liczb z rozkładu Gaussa

**Proszę:**

Skorzystać z centralnego twierdzenia granicznego i wygenerować macierz `z` zawierającą próbkę 10 000 liczb pochodzącących z rozkładu normalnego. Algorytm jest następujący:

1) generacja `nComponents` liczb $x_{i}$ z rozkładu płaskiego `[0,1]`
2) $z = \sum_{i} x_{i} - 0.5 \cdot$ nComponents

Korzystając z powyższego algorytmu proszę wygenerować macierz numpy zawierającą`nSamples=10000` liczb z rozkładu normalnego

**Uwaga:** nie wolno używać pętli! Kod generacji może zająć maksymalnie **4** linie w tym dwie linie postaci:
```Python

nComponents = 12
```

````python
%%time
nSamples = 10000
nComponents = 12
x = np.random.default_rng().uniform(0.0, 1.0, (nComponents, nSamples))
z = np.sum(x, axis = 0) - 0.5*nComponents
pass
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def quadraticEqGenerator(nSamples):
    return np.random.default_rng().uniform(-1, 1, (nSamples, 3))

def quadraticEqSolution(coeff):
    a, b, c = coeff[:, 0], coeff[:, 1], coeff[:, 2]
    delta = b**2 - 4*a*c
    has_solutions = delta >= 0
    x1, x2 = np.full_like(a, np.nan), np.full_like(a, np.nan)
    x1[has_solutions] = (-b[has_solutions] + np.sqrt(delta[has_solutions]))/(2*a[has_solutions])
    x2[has_solutions] = (-b[has_solutions] - np.sqrt(delta[has_solutions]))/(2*a[has_solutions])
    solutions = np.column_stack((x1, x2))
    return solutions

def plotQuadraticEqSolvability(data, interactive=False):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    solutions = quadraticEqSolution(data)
    solvable = ~np.isnan(solutions[:, 0])
    ax.scatter(data[:, 0][solvable], data[:, 1][solvable], data[:, 2][solvable], c='g', label='Solvable')
    ax.scatter(data[:, 0][~solvable], data[:, 1][~solvable], data[:, 2][~solvable], c='r', label='Unsolvable')

    ax.set_xlabel('a')
    ax.set_ylabel('b')
    ax.set_zlabel('c')
    ax.legend()

    if interactive:
        plt.show()
    else:
        plt.savefig("quadratic_solvability.png")
        plt.close()

def main():
    nSamples = 1000
    data = quadraticEqGenerator(nSamples)
    plotQuadraticEqSolvability(data, interactive=True)

main()
````
